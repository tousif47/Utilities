```python
# You will see a WARNING message in the terminal, ignore it. It's not critical.

# Import necessary libraries
from moviepy.editor import VideoFileClip
from google.colab import files
import os

# Function to convert .webm to .mp4
def convert_webm_to_mp4(input_path, output_path):
    # Load the .webm video
    video = VideoFileClip(input_path)

    # Write the video to .mp4 format
    video.write_videofile(output_path, codec='libx264', audio_codec='aac')

    # Close the video file
    video.close()

# Upload the .webm file
uploaded = files.upload()

# Get the uploaded file name
uploaded_file_name = list(uploaded.keys())[0]

# Define the output file name
output_file_name = uploaded_file_name.replace('.webm', '.mp4')

# Convert the uploaded .webm file to .mp4
convert_webm_to_mp4(uploaded_file_name, output_file_name)

# Download the converted .mp4 file to the local machine
files.download(output_file_name)
```
